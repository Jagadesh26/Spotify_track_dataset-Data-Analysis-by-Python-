# finding percentage of streams 

```python
s=sp_data['streams']/sp_data['streams'].sum()*100
s_streams=s.round(3).sort_values(ascending=False)
s_streams
```

# adding new column for percentage streams

```python
s=sp_data['streams']/sp_data['streams'].sum()*100
sp_data['percentage_of_streams']=s.round(3)
```

```python
sp_data
```

```python
sp_data['percentage_of_streams'].describe()
```

# based on percentage streams sperate the catagory of track in three type Super hit,hit,average


```python
sp_data.loc[(sp_data['percentage_of_streams']<1) & (sp_data['percentage_of_streams']>=0.13),'track_status']='Super Hit'
sp_data.loc[(sp_data['percentage_of_streams']<0.129) & (sp_data['percentage_of_streams']>=0.05),'track_status']='Hit'
sp_data.loc[sp_data['percentage_of_streams']<=0.049,'track_status']='Average'
```
